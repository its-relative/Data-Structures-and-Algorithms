Explanation

### we just need to put rjust

 which shifts the whole pattern or should i say the symbols in a line towards right by n if used like

 ```py
 has = "#"*i
 print(has.rjust(n))
 ```

````python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'staircase' function below.
#
# The function accepts INTEGER n as parameter.
#

def staircase(n):
    # Write your code here
    for i in range(1,n+1):
        has = "#"*(i)
        # spc = " "*(n -1-i+1)
        print(has.rjust(n))
        
        # print(' '*(n-i),"#"*(i))
if __name__ == '__main__':
    n = int(input().strip())

    staircase(n)
````
